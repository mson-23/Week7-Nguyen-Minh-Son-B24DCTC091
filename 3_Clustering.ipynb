```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("student-mat.csv", sep=";")
```

```python
# Chọn đặc trưng (Features) để gom cụm
X = df[["G1", "G2", "G3"]]

# Khởi tạo mô hình KMeans và phân lớp dữ liệu
kmeans = KMeans(n_clusters=3, random_state=42)

# Gán nhãn cụm vừa tìm được vào lại một cột mới trong df
df["cluster"] = kmeans.fit_predict(X)
```

```python
# Trực quan hóa kết quả bằng biểu đồ phân tán (Scatter plot)
plt.scatter(df["G1"], df["G2"], c=df["cluster"])
plt.xlabel("G1")
plt.ylabel("G2")
plt.title("Student Clusters")
plt.show()
```
