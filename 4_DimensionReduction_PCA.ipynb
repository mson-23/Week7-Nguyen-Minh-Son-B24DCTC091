```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Đọc dữ liệu và in ra 5 dòng đầu tiên xem thử
df = pd.read_csv("student-mat.csv", sep=";")
df.head()
```

```python
# Lựa chọn các cột đặc trưng (Features)
X = df[["G1", "G2", "G3", "studytime", "failures", "absences"]]

# Bước vô cùng quan trọng đối với PCA: Chuẩn hoá dữ liệu (Scale)
X_scaled = StandardScaler().fit_transform(X)

# Tạo và huấn luyện mô hình PCA (Giảm số lượng đặc trưng xuống còn 2)
pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)
```

```python
# Lấy cột giá trị thứ 1 làm trục x, cột giá trị thứ 2 làm trục y
plt.scatter(components[:, 0], components[:, 1])

plt.title("PCA Visualization")
plt.show()
```
