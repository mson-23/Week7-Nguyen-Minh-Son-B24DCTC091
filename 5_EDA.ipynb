```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student-mat.csv", sep=";")
```

```python
# Phân tích thống kê (giá trị trung bình, độ lệch chuẩn, min, max...)
print(df.describe())

# Đếm tổng số ô bị trống/lỗi trong từng cột
print(df.isnull().sum())
```

```python
plt.hist(df["G3"], bins=20)
plt.title("Phân bổ Final Grade (G3)")
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["G1"], df["G3"])
plt.xlabel("G1")
plt.ylabel("G3")
plt.title("G1 vs Final Grade")
plt.show()
```
